```python
class Employer:
    def __init__(self, name, id):
        self.__name = name
        self.__id = id

    def get_info(self):
        return f"Имя: {self.__name}, ID номер: {self.__id}"


class Manager(Employer):
    def __init__(self, name, id, department):
        Employer.__init__(self, name, id)
        self.departament = department

    def manage_project(self):
        return "я имею право управлять проектами"

    def get_info(self):
        return f"{Employer.get_info(self)}, я заведую департментом: {self.departament}"

    def get_departament(self):
        return self.departament


class Technician(Employer):
    def __init__(self, name, id, spec):
        Employer.__init__(self, name, id)
        self.spec = spec

    def get_info(self):
        return f"{Employer.get_info(self)}, моя специализация: {self.spec}"

    def get_spec(self):
        return self.spec

    def perform_maintenance(self):
        return "Я провожу тех обслуживание"


class TechManager(Manager, Technician):
    def __init__(self, name, id, spec, department):
        Employer.__init__(self, name, id)
        Manager.__init__(self, name, id, department)
        Technician.__init__(self, name, id, spec)
        self.emp_lst = []

    def get_info(self):
        return f"{Employer.get_info(self)}, я заведую отделом: {Manager.get_departament(self)}, моя специализация: {Technician.get_spec(self)}"

    def perform_maintenance(self):
        return Technician.perform_maintenance(self)

    def manage_project(self):
        return "и я тоже имею на это право"

    def add_employer(self, employer):
            self.emp_lst.append(employer)

    def get_info_team(self):
        return f"список рабочих: {self.emp_lst}"


Employer_1 = Employer(name="Lalatina", id="0001")
Employer_2 = Employer(name="Evgen", id="0002")
Employer_3 = Manager(name="Sergey", id="0003", department="Разработка игр")
Employer_4 = Technician(name="Itan", id="0004", spec="Электрик")
Employer_5 = TechManager(name="Ryan", id="0005", department="Разработка Приложений", spec="Инженер")

Employer_5.add_employer(Employer_1.get_info())
Employer_5.add_employer(Employer_2.get_info())



print(Employer_1.get_info())
print(Employer_2.get_info())
print(Employer_3.get_info())
print(Employer_3.manage_project())
print(Employer_4.get_info())
print(Employer_5.get_info())
print(Employer_5.get_info_team())
```
